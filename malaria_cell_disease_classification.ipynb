```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset_path = "./"
images = {}
target_size = (100, 100)

X = []
y = []

for class_name in os.listdir(dataset_path):
    class_dir = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_dir):
        images[class_name] = []

        for img_name in os.listdir(class_dir):
            img_path = os.path.join(class_dir, img_name)

            try:
                with Image.open(img_path) as img:
                    img = img.convert("RGB").resize(target_size)
                    img_array = np.array(img) / 255.0
                    X.append(img_array)
                    y.append(class_name)
                    images[class_name].append(img.copy())
            except Exception as e:
                print(f"Could not open {img_path}: {e}")
```

```python
images
```

```python
# Display a few images from each class
for class_name, imgs in images.items():
    print(f"Class: {class_name}")
    plt.figure(figsize=(10, 5))
    for i, img in enumerate(imgs[:5]):  # Show up to 5 images per class
        plt.subplot(1, 5, i+1)
        plt.imshow(img)
        plt.axis('off')
    plt.show()
```

```python
# Parasitized = 0
# Uninfected = 1
y = np.array(pd.get_dummies(y, drop_first=True, dtype=float))
y
```

# Model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, shuffle=True, stratify=y)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
```

```python
model = keras.Sequential()
# CNN
model.add(keras.Input(shape=(100, 100, 3)))
model.add(layers.Conv2D(filters=32, kernel_size=(3, 3), activation="relu")) # kernel_size is the filter size (3, 3) in this case
model.add(layers.MaxPool2D(pool_size=(2, 2)))
model.add(layers.Conv2D(filters=64, kernel_size=(3, 3), activation="relu"))
model.add(layers.MaxPool2D(pool_size=(2, 2)))
# Classification
model.add(layers.Flatten())
model.add(layers.Dense(64, activation="relu"))
model.add(layers.Dense(1, activation="sigmoid"))

print(model.summary())
```

```python
# Setup the model configuration
model.compile(
    loss=keras.losses.BinaryCrossentropy(),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"]
)
```

```python
# Make the model actualy train
model.fit(X_train, y_train, batch_size=32, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
# Save
model.save("malaria_model.h5")
```
